## while문

```python
#while문 만들어보기

i=0
while i < 10 :
    print("성장하는 괴물")
    i += 1
```

```python
#3.while문 조건식 증가, 감소

#1.기본적인 증가 유형

i=1
while i < 10:
    print('이정도로 성장할줄이야',i)
    i+=1

#기본적인 감소 유형
i = 10
while i > 0 :
    print('이정도로 성장할줄이야',i)
    i-=1
```

```python
#4.while문을 조심스럽게 사용해야하는 이유
#조건이 잘못 설정되면 무한 루프에 빠져 과도한 리소스 소모
#하지만, 잘 사용하면 for문보다 효율적


# while True:
#     print(1)

while True:
    print('내가 산 주식 폭등')
```

```python
#5.while문이 for보다 강점인 부분
#무수히 많이 실행되어야 할때

import random

# random.randint(1,6)

i = 0
while i !=4:
    i = random.randint(1,6)
    print(i)
```

```python
#6. while 문에도 else가 있어요

양파 = 10
while 양파>10:
    print("양파",양파,"개 남았습니다")

    if 양파 == 0:
        break
    양파 -=1
else:
    print("양파 다 팔림")
```

```python
#7. 반복문에 흐름을 제어하는 continue와 break

i = 0

while i < 10:
    i+=1
    if i % 2 ==0:
        continue
    print(i)


for i in range(10):
    if i %2 == 0:
        continue
    print(i)
```

```python
#9. 예제로 while문 익히기

#콜라츠하는 사람은 주어진 수가 1이 될때까지 다음 작업을 반복하면, 모든 수를 1로 만들수 있다고 추측했다
#1-1. 입력수가 짝수라면 나누기 2
#1-2. 입력수가 홀수라면 3을 곱하고 1을 더한다.
#3. 위 결과로 나온 수가 같은 작업을 1이 될때까지 반복
#입력수가 6이라면 6->3->10->5->16->8->4->2->1

#num변수에 있는 수를 1로 만드는데 몇번이나 작업을 반복하는지 출력
num = 7

anser = 0

while num != 1:
    if num % 2 == 0:
        num = num/2
    else :
        num = num*3 +1

anser += 1

print(anser)
```
